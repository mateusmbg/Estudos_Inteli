# Ponderada

## Enunciado da Ponderada

Suponha que uma empresa de varejo esteja tentando determinar quais produtos devem ser mantidos em estoque para atender  demanda dos clientes. A empresa tem histricos de vendas dirias dos ltimos 30 dias para cada produto e deseja usar a entropia para determinar a importncia relativa de cada produto.

Os dados de vendas dirias para cada produto esto na tabela a seguir:

**Produto/Vendas Dirias:**

A/ 10<br>
B/ 5<br>
C/ 3<br>
D/ 2<br>
E/ 0<br>
F/ 0<br>

Calcule a entropia desse sistema.

## Resposta

A entropia mxima em bits desse sistema  de 2, isso porque existem 4 variveis que vo entrar dentro do clculo da entropia. A partir da separao das varveis e do valor das mesmas  possv el iniciar o clculo e resoluo do exerccio.

### Exerccio Resolvido no Papel!


```python
def calcular_entropia(prob) :
    return -sum(p * math.log2(p) for p in prob if p > 0)
```

```python
# Probabilidades das vendas
p_a = 0.5
p_b = 0.25
p_c = 0.15
p_d = 0.1

entropia = calcular_entropia([p_a, p_b, p_c, p_d])
print(f"Entropia das vendas dirias: {entropia} bits")
```
