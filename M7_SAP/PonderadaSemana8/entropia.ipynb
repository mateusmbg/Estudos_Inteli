# Ponderada

## Enunciado da Ponderada

Suponha que uma empresa de varejo esteja tentando determinar quais produtos devem ser mantidos em estoque para atender à demanda dos clientes. A empresa tem históricos de vendas diárias dos últimos 30 dias para cada produto e deseja usar a entropia para determinar a importância relativa de cada produto.

Os dados de vendas diárias para cada produto estão na tabela a seguir:

**Produto/Vendas Diárias:**

A/ 10<br>
B/ 5<br>
C/ 3<br>
D/ 2<br>
E/ 0<br>
F/ 0<br>

Calcule a entropia desse sistema.

## Resposta

A entropia máxima em bits desse sistema é de 2, isso porque existem 4 variáveis que vão entrar dentro do cálculo da entropia. A partir da separação das varáveis e do valor das mesmas é possív el iniciar o cálculo e resolução do exercício.

### Exercício Resolvido no Papel!


```python
def calcular_entropia(prob) :
    return -sum(p * math.log2(p) for p in prob if p > 0)
```

```python
# Probabilidades das vendas
p_a = 0.5
p_b = 0.25
p_c = 0.15
p_d = 0.1

entropia = calcular_entropia([p_a, p_b, p_c, p_d])
print(f"Entropia das vendas diárias: {entropia} bits")
```

### Cálculo do Log

```python
# Aqui é feito o cálculo do log que utilizei para realizar a ponderada no papel
print(math.log2(p_b))
print(math.log2(p_c))
print(math.log2(p_d))
```
